# Python Data Structures for Data Mining
This notebook provides a foundational overview of the key Python data structures used in data mining, ranging from built-in types to advanced libraries like NumPy and Pandas.

## 1. Python Lists
Lists are mutable sequences used to store collections of data. They are the starting point for handling raw data tuples.

```python
# 1. Standard List & List Comprehension
student_marks = [73, 79, 70, 55, 85.5]

# List Comprehension: Memberi bonus 5 poin ke semua siswa (Sangat umum di data mining)
boosted_marks = [mark + 5 for mark in student_marks]
print(f"Nilai setelah bonus: {boosted_marks}")
```

## 2. Multidimensional Lists
A "list of lists" is often used to represent matrices or tabular data in pure Python.

```python


# 2. Multidimensional List (Matriks Mentah)
# Format: [ID, Umur, Nilai]
data_matrix = [
    [1, 20, 85],
    [2, 22, 90],
    [3, 21, 78]
]

# Cara membaca data_matrix[row][column]:
# Row 1 (Indeks 1) adalah siswa ke-2. Column 2 (Indeks 1) adalah Umur.
second_student_age = data_matrix[1][1]
print(f"Umur Siswa Kedua: {second_student_age}")

# Menambahkan 'Baris' (Siswa baru) ke dalam dataset mentah
new_student = [4, 23, 88]
data_matrix.append(new_student)
```

### Tabel Metode Utama List untuk Data Mining

| Metode | Kegunaan dalam Data Mining | Contoh Kode |
| :--- | :--- | :--- |
| **`.append()`** | Menambahkan satu data baru (baris baru) ke akhir list. | `data.append([4, 22, 90])` |
| **`.extend()`** | Menggabungkan dua list menjadi satu (menggabungkan dataset). | `list1.extend(list2)` |
| **`.sort()`** | Mengurutkan data (misal: mencari nilai tertinggi ke terendah). | `marks.sort(reverse=True)` |
| **`.count()`** | Menghitung kemunculan nilai tertentu (frekuensi sederhana). | `count_80 = marks.count(80)` |
| **`len()`** | Menghitung jumlah total baris/data dalam list. | `total_data = len(data_matrix)` |
| **`sum()`** | Menjumlahkan seluruh angka dalam list (untuk hitung rata-rata manual). | `total_score = sum(marks)` |
| **Slicing `[:]`** | Mengambil sebagian data (misal: memisahkan data training dan testing). | `train_data = data[:80]` |

## 3. Dictionaries
Dictionaries store data in **Key-Value** pairs, making them ideal for structured records.

```python
patient = {
    "age": 45,
    "sex": "M",
    "cholesterol": 240,
    "target": 1
}

print(f"Patient Age: {patient['age']}")

# Updating a value
patient["cholesterol"] = 235
print(f"Updated Record: {patient}")
```

```python
patient = {
    "id": "PT-90210",
    "demographics": {
        "age": 45,
        "sex": "M",
    },
    "vitals": {
        "cholesterol": 240,
        "blood_pressure": "130/85",
        "bmi": 28.5
    },
    "history": ["hypertension", "smoker"],
    "target": 1  # 1 = Risk, 0 = No Risk
}

# Safer way to access nested data
age = patient["demographics"].get("age", "Unknown")
glucose = patient["vitals"].get("glucose", 0.0) # Returns 0.0 if missing

print(f"Patient {patient['id']} Age: {age}")
print(f"Glucose Level: {glucose}")
```

```python
# A collection of patients
patients_db = [
    patient, # Our first patient
    {
        "id": "PT-11550",
        "demographics": {"age": 32, "sex": "F"},
        "vitals": {"cholesterol": 190, "blood_pressure": "115/75", "bmi": 22.1},
        "history": [],
        "target": 0
    }
]

# Example: Average age calculation
avg_age = sum(p["demographics"]["age"] for p in patients_db) / len(patients_db)
print(f"Dataset Average Age: {avg_age}")
```

### Tabel Metode Utama Dictionary untuk Data Mining

| Metode | Kegunaan dalam Data Mining | Contoh Kode |
| :--- | :--- | :--- |
| **`.get()`** | Mengambil nilai secara aman tanpa risiko error jika *key* tidak ditemukan (penting untuk data yang tidak lengkap). | `val = patient.get("bmi", 0.0)` |
| **`.update()`** | Memperbarui beberapa fitur (kolom) sekaligus dalam satu rekaman data. | `patient.update({"target": 0, "cholesterol": 210})` |
| **`.keys()`** | Mendapatkan daftar semua nama fitur atau atribut dalam dataset. | `features = patient.keys()` |
| **`.values()`** | Mengambil semua nilai data untuk diproses secara numerik. | `data_points = patient.values()` |
| **`.items()`** | Melakukan iterasi melalui pasangan fitur-nilai untuk transformasi data. | `for k, v in patient.items(): print(k, v)` |
| **`.pop()`** | Menghapus kolom yang tidak relevan atau bersifat sensitif (seperti ID atau Nama). | `patient.pop("id")` |
| **`.clear()`** | Menghapus seluruh data dalam rekaman untuk mengosongkan memori. | `patient.clear()` |

## 4. NumPy Arrays
NumPy arrays are designed for fast numerical computations and support vectorized operations.

```python
import numpy as np

# 1. Array 1D & Filtering (Boolean Indexing)
# Mencari data tertentu tanpa menggunakan 'for loop'
marks = np.array([80, 90, 70, 60, 95])
top_marks = marks[marks >= 85]  # Filter instan: ambil yang >= 85
print(f"Nilai di atas 85: {top_marks}")

# 2. Agregasi Berdasarkan Axis (Penting untuk Data Tabular)
# Baris: Mahasiswa, Kolom: Nilai Ujian 1 & Ujian 2
data_ujian = np.array([
    [80, 85], 
    [90, 88], 
    [70, 75]
])

# axis=0 adalah kolom (rata-rata per mata kuliah)
# axis=1 adalah baris (rata-rata per mahasiswa)
mean_per_student = np.mean(data_ujian, axis=1)
print(f"Rata-rata per mahasiswa: {mean_per_student}")

# 3. Reshaping (Menyiapkan data untuk Machine Learning)
# Mengubah array 1D menjadi matriks kolom (sering diminta oleh Scikit-Learn)
reshaped_data = marks.reshape(-1, 1)
print(f"Bentuk data setelah reshape:\n{reshaped_data.shape}")
```

### Tabel Metode Utama NumPy untuk Data Mining

| Metode | Kegunaan dalam Data Mining | Contoh Kode |
| :--- | :--- | :--- |
| **`np.mean()`** | Menghitung rata-rata (pusat data) pada baris atau kolom tertentu. | `np.mean(data, axis=0)` |
| **`np.std()`** | Menghitung standar deviasi untuk melihat sebaran/variansi data. | `deviation = np.std(data)` |
| **`.reshape()`** | Mengubah dimensi data tanpa mengubah isinya (misal: dari 1D ke 2D). | `data.reshape(5, 1)` |
| **`np.where()`** | Logika "If-Else" versi cepat untuk mengubah nilai berdasarkan kondisi. | `np.where(data > 70, "Lulus", "Gagal")` |
| **`.T` (Transpose)** | Menukar baris menjadi kolom (berguna dalam operasi aljabar linier). | `matrix_rotated = data.T` |
| **`np.argmax()`** | Menemukan indeks dengan nilai tertinggi (sering digunakan untuk prediksi kelas). | `best_class = np.argmax(predictions)` |
| **`np.nan_to_num()`** | Mengganti nilai kosong (NaN) dengan angka 0 atau nilai lainnya. | `clean_data = np.nan_to_num(data)` |

## 5. Pandas DataFrames
The DataFrame is the standard for tabular data manipulation, providing Excel-like functionality via code.

```python
import pandas as pd
import numpy as np

# 1. Membuat DataFrame dengan data yang lebih kompleks (ada nilai NaN/Kosong)
data = {
    'Student': ['S1', 'S2', 'S3', 'S4', 'S5'],
    'Age': [21, 19, 22, 20, np.nan], # Ada data hilang
    'Mark': [85, 90, 78, 92, 88],
    'Class': ['A', 'B', 'A', 'B', 'A']
}
df = pd.DataFrame(data)

# 2. Data Cleaning: Mengisi nilai Age yang kosong dengan rata-rata
df['Age'] = df['Age'].fillna(df['Age'].mean())

# 3. Feature Engineering: Menambah kolom status berdasarkan nilai
df['Status'] = np.where(df['Mark'] >= 80, 'High', 'Standard')

# 4. Aggregation: Menghitung rata-rata nilai per kelas (Penting untuk pola data)
class_summary = df.groupby('Class')['Mark'].mean()

print("--- DataFrame Setelah Cleaning ---")
print(df)
print("\n--- Rata-rata Nilai per Kelas ---")
print(class_summary)

# 5. Statistik Deskriptif: Ringkasan cepat seluruh data numerik
print("\n--- Ringkasan Statistik ---")
print(df.describe())
```

### Tabel Metode Utama Pandas untuk Data Mining

| Metode | Kegunaan dalam Data Mining | Contoh Kode |
| :--- | :--- | :--- |
| **`.info()`** | Melihat tipe data dan mendeteksi jumlah nilai kosong (null) di setiap kolom. | `df.info()` |
| **`.describe()`** | Mendapatkan statistik lengkap (mean, median, min, max) untuk analisis awal. | `df.describe()` |
| **`.fillna()`** | Menangani *missing values* dengan mengisi data kosong menggunakan nilai tertentu. | `df['col'].fillna(0)` |
| **`.groupby()`** | Mengelompokkan data berdasarkan kategori untuk melihat pola antar grup. | `df.groupby('kategori').mean()` |
| **`.value_counts()`** | Menghitung frekuensi kemunculan data (sangat berguna untuk fitur kategori). | `df['gender'].value_counts()` |
| **`.merge()`** | Menggabungkan dua dataset berbeda berdasarkan kunci tertentu (seperti JOIN di SQL). | `pd.merge(df1, df2, on='id')` |
| **`.apply()`** | Menerapkan fungsi kustom pada setiap baris atau kolom secara efisien. | `df['val'].apply(lambda x: x*2)` |
