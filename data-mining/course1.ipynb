## 1. Python Lists
Lists are mutable sequences used to store collections of data. They are the starting point for handling raw data tuples.

```python
# 1. Standard List & List Comprehension
student_marks = [73, 79, 70, 55, 85.5]

# List Comprehension: Memberi bonus 5 poin ke semua siswa (Sangat umum di data mining)
boosted_marks = [mark + 5 for mark in student_marks]
print(f"Nilai setelah bonus: {boosted_marks}", "type", type(boosted_marks))
print(f"type item dari list comprehension: {[type(mark) for mark in boosted_marks]}")
```

```python
# indexing: Mengakses nilai siswa kedua (index dimulai dari 0)
print(student_marks[1])
# negative indexing: Mengakses nilai siswa terakhir
print(student_marks[-2])
# slicing: Mengakses nilai siswa kedua hingga keempat
print(student_marks[1:4])
# negative slicing: Mengakses nilai siswa pertama hingga ketiga
print(student_marks[-3:-1])
print(student_marks[:3])  # Mengakses nilai siswa pertama hingga ketiga (slicing tanpa start index)
print(student_marks[2:])  # Mengakses nilai siswa ketiga hingga akhir (slicing tanpa end index)

print(student_marks[::-1]) # Mengakses nilai siswa dalam urutan terbalik (slicing dengan step -1)
print(student_marks[::2])  # Mengakses nilai siswa dengan step 2 (slicing dengan step 2)
print(student_marks[1::2]) # Mengakses nilai siswa dengan step 2 dimulai dari index 1 (slicing dengan step 2 dan start index 1)
```

```python
# 2. String Indexing & Slicing
name = "Data Mining"
print(name[0])  # Mengakses karakter pertama
print(name[::2])  # Mengakses karakter dengan step 2 (slicing string dengan step 2)
```

## 2. Multidimensional Lists
A "list of lists" is often used to represent matrices or tabular data in pure Python.

```python


# 2. Multidimensional List (Matriks Mentah)
# Format: [ID, Umur, Nilai]
data_matrix = [
    [1, 20, 85], # Siswa pertama indeks 0
    [2, 22, 90], # Siswa kedua  indeks 1
    [3, 21, 78] # Siswa ketiga  indeks 2
    # ID indeks 0, Umur indeks 1, Nilai indeks 2
]

# Cara membaca data_matrix[row][column]:
# Row 1 (Indeks 1) adalah siswa ke-2. Column 2 (Indeks 1) adalah Umur.
second_student_age = data_matrix[1][1]
print(f"Umur Siswa Kedua: {second_student_age}")
print(f"Umur Siswa Ketiga: {data_matrix[2][1]}")
print(f"Nilai Siswa Kedua: {data_matrix[1][2]}")
print(f"Nilai Siswa Ketiga: {data_matrix[2][2]}")

# sliceing pada matriks mentah:
print(f"Data Siswa Pertama dan Kedua: {data_matrix[0:2]}")
print(f"Data umur dan nilai Siswa Kedua: {data_matrix[1][1:3]}")  # Mengakses umur dan nilai siswa kedua

# Menambahkan 'Baris' (Siswa baru) ke dalam dataset mentah
new_student = [4, 23, 88]
data_matrix.append(new_student)
print(f"Data Matrix setelah menambahkan siswa baru: {data_matrix}")
```

## 3. Dictionaries
Dictionaries store data in **Key-Value** pairs, making them ideal for structured records.

```python
patient = {
    "age": 45,
    "sex": "M",
    "cholesterol": 240,
    "target": 1
}

print(f"Patient Age: {patient['age']}")

# Updating a value
patient["cholesterol"] = 235
print(f"Updated Record: {patient}")
```

```python
# nested dictionary untuk data pasien yang lebih kompleks
patient = {
    "id": "PT-90210",
    "demographics": {
        "name": "John Doe",
        "age": 45,
        "sex": "M",
    },
    "vitals": {
        "cholesterol": 240,
        "blood_pressure": "130/85",
        "bmi": 28.5,
        "glucose": 110.0
    },
    "history": ["hypertension", "smoker"],
    "target": 1  # 1 = Risk, 0 = No Risk
}

# Safer way to access nested data
age = patient["demographics"].get("age", "Unknown")
glucose = patient["vitals"].get("glucose", "0") # apabila kita tidak yakin key tersebut ada, kita bisa menggunakan get() dengan default value

print(f"Patient {patient['demographics']['name']}, Age: {age}") # apabila kita yakin key tersebut ada, kita bisa langsung akses tanpa get()
print(f"Glucose Level: {glucose}")
```

```python
# A collection of patients
patients_db = [
    patient, # Our first patient
    {
        "id": "PT-11550",
        "demographics": {"name": "Jane Smith", "age": 32, "sex": "F"},
        "vitals": {"cholesterol": 190, "blood_pressure": "115/75", "bmi": 22.1},
        "history": [],
        "target": 0
    }
]

# Example: Average age calculation
avg_age = sum(p["demographics"]["age"] for p in patients_db) / len(patients_db)
print(f"Dataset Average Age: {avg_age}")
print(patients_db)
```

## 4. NumPy Arrays
NumPy arrays are designed for fast numerical computations and support vectorized operations.

```python
import numpy as np

# 1. Array 1D & Filtering (Boolean Indexing)
# Mencari data tertentu tanpa menggunakan 'for loop'
marks = np.array([80, 90, 70, 60, 95])
top_marks = marks[marks >= 85]  # Filter instan: ambil yang >= 85
print(f"Nilai di atas 85: {top_marks}")

# 2. Agregasi Berdasarkan Axis (Penting untuk Data Tabular)
# Baris: Mahasiswa, Kolom: Nilai Ujian 1 & Ujian 2
data_ujian = np.array([
    [80, 90], 
    [50, 50], 
    [70, 90]
])

# axis=0 adalah kolom (rata-rata per mata kuliah)
# axis=1 adalah baris (rata-rata per mahasiswa)
mean_per_student = np.mean(data_ujian, axis=1)
print(f"Rata-rata per mahasiswa: {mean_per_student}")

# 3. Reshaping (Menyiapkan data untuk Machine Learning)
# Mengubah array 1D menjadi matriks kolom (sering diminta oleh Scikit-Learn)
reshaped_data = marks.reshape(-1, 1)
print(f"Bentuk data setelah reshape:\n{reshaped_data.shape}")
print("data\n", data_ujian, "type", type(data_ujian), "dimension")
```

## 5. Pandas DataFrames
The DataFrame is the standard for tabular data manipulation, providing Excel-like functionality via code.

```python
import pandas as pd
import numpy as np

# 1. Membuat DataFrame dengan data yang lebih kompleks (ada nilai NaN/Kosong)
data = {
    'Studentid': ['S1', 'S2', 'S3', 'S4', 'S5'],
    'Age': [21, 19, 22, 20, np.nan], # Ada data hilang
    'Mark': [85, 90, 78, 92, 40], # 30%
    'UTS': [85, 90, 78, 92, 40], # 30%
    'UAS': [85, 90, 78, 70, 80], # 50%
    'Present': [10, 14, 14, 13, 10], # max 14 20%
    'Class': ['A', 'B', 'A', 'B', 'A']
}
df = pd.DataFrame(data)
print("--- DataFrame Sebelum Cleaning ---")
print(df)
# 2. Data Cleaning: Mengisi nilai Age yang kosong dengan rata-rata
df['Age'] = df['Age'].fillna(df['Age'].mean())

def grade_count(x):
    if x>= 80:
        return "A"
    elif x < 50:
        return "E"
    elif 50 <= x <65:
        return "D"
    elif 65 <= x <75:
        return "C"
    else:
        return "B"
    
def calculate_total(present, uts, uas):

    present = present / 14 * 20 # 20%
    uts = uts / 100 * 30 # 30%
    uas = uas / 100 * 50 # 50%
    
    return present + uts + uas

# 3. Feature Engineering: Menambah kolom status berdasarkan nilai
df['Grade'] = df['Mark'].apply(lambda x: grade_count(x))
df['Status'] = np.where(df['Mark'] >= 75, 'Lulus', 'Tidak Lulus')
df['student_number'] = df['Studentid'].apply(lambda x: x.replace('S', '001-0'))

# implement calculate_total() into df['total] column
# df['total'] = 
df = df.drop(columns=['Studentid'])

# 4. Aggregation: Menghitung rata-rata nilai per kelas (Penting untuk pola data)
class_summary = df.groupby('Class')['Mark'].mean()

print("--- DataFrame Setelah Cleaning ---")
print(df)
print("\n--- Rata-rata Nilai per Kelas ---")
print(class_summary)

# 5. Statistik Deskriptif: Ringkasan cepat seluruh data numerik
print("\n--- Ringkasan Statistik ---")
print(df.describe())
```
