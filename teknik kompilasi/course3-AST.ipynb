# Memahami Abstract Syntax Tree (AST)

**Abstract Syntax Tree (AST)** adalah struktur data pohon yang mewakili struktur sintaksis abstrak dari kode sumber yang ditulis dalam bahasa pemrograman. 

AST berfungsi sebagai **Intermediate Representation (IR)** atau representasi antara yang sangat penting bagi interpreter dan kompiler untuk memahami esensi dari instruksi pemrograman tanpa harus memproses setiap detail kecil seperti tanda kurung atau aturan tata bahasa yang redundan.

## 1. Perbedaan AST dan Parse Tree (Concrete Syntax Tree)

Meskipun keduanya berbentuk pohon, AST memiliki karakteristik yang berbeda dari *Parse Tree*:

- **AST lebih padat**: AST hanya menangkap "esensi" dari input, sedangkan *Parse Tree* mencatat setiap aturan produksi gramatika yang diterapkan oleh parser.
- **Abstrak**: AST tidak menyimpan simbol-simbol teknis seperti tanda kurung atau titik koma karena makna operasionalnya sudah diwakili oleh struktur pohon itu sendiri.
- **Operator sebagai Node**: Dalam AST, operator (seperti `+` atau `*`) biasanya menjadi node akar atau node interior, sementara operan (seperti angka) menjadi anak-anaknya.



## 2. Representasi Prioritas Operator dalam AST

Dalam AST, **operator precedence** (prioritas operator) ditentukan oleh posisi node dalam hierarki pohon. 

Operator dengan prioritas lebih tinggi (yang harus dikerjakan lebih dulu) akan diletakkan pada level yang **lebih rendah** atau lebih dalam di dalam pohon dibandingkan operator dengan prioritas rendah. 

**Contoh:** Dalam ekspresi `7 + 3 * 2`, operasi perkalian (`*`) akan berada di sub-pohon yang lebih dalam daripada penjumlahan (`+`).

## 3. Implementasi Node AST di Python

Berikut adalah implementasi dasar kelas-kelas untuk mewakili berbagai jenis node dalam sebuah pohon.

```python
class AST:
    """Base class untuk semua node AST."""
    pass

class BinOp(AST):
    """Node untuk operasi biner (seperti +, -, *, /)."""
    def __init__(self, left, op, right):
        self.left = left    # Node operan kiri
        self.op = op        # Token atau simbol operator
        self.right = right  # Node operan kanan

class Num(AST):
    """Node untuk mewakili angka tunggal."""
    def __init__(self, value):
        self.value = value

print("Kelas AST berhasil didefinisikan.")
```

## 4. Evaluasi AST Menggunakan Visitor Pattern

Untuk mengeksekusi kode dari AST, kita menggunakan metode **Postorder Traversal**. Di Python, pola ini sering diimplementasikan menggunakan *Visitor Pattern*.

```python
class NodeVisitor:
    def visit(self, node):
        method_name = 'visit_' + type(node).__name__
        visitor = getattr(self, method_name, self.generic_visit)
        return visitor(node)

    def generic_visit(self, node):
        raise Exception(f'Tidak ada metode visit_{type(node).__name__}')

class Interpreter(NodeVisitor):
    def visit_BinOp(self, node):
        if node.op == '+':
            return self.visit(node.left) + self.visit(node.right)
        elif node.op == '*':
            return self.visit(node.left) * self.visit(node.right)
        elif node.op == '-':
            return self.visit(node.left) - self.visit(node.right)
        elif node.op == '/':
            return self.visit(node.left) / self.visit(node.right)

    def visit_Num(self, node):
        return node.value

# Simulasi pembuatan pohon untuk ekspresi: 7 + 3 * 2
# Struktur: (7 + (3 * 2))
tree = BinOp(
    left=Num(7),
    op='+',
    right=BinOp(left=Num(3), op='*', right=Num(2))
)

interpreter = Interpreter()
hasil = interpreter.visit(tree)
print(f"Hasil evaluasi AST (7 + 3 * 2) adalah: {hasil}")
```

## 5. Manfaat Menggunakan AST

1. **Optimasi Kode**: Kompiler dapat melakukan transformasi kode (seperti *constant folding*) pada tingkat pohon ini.
2. **Analisis Semantik**: Memudahkan pengecekan tipe data dan validasi apakah variabel sudah didefinisikan.
3. **Generasi Kode**: Memudahkan pembuatan *Three Address Code* atau bahasa mesin target dari struktur yang sudah bersih.
